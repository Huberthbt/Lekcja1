```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('himeji.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.title("Oryginalny obraz")
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.title("W skali szroci")
plt.imshow(img_gray, cmap= 'gray')
```

```python
print(img_gray)
```

```python
import numpy as np

def apply_filter(img, filter):
  height, width = img.shape
  filtered_img = np.zeros((height - 2, width - 2), dtype=np.float32)

  for y in range(1, height - 1):

    for x in range(1, width - 1):
      neighborhood = img[y - 1 : y + 2, x - 1 : x + 2]
      convolutino_value = np.sum(neighborhood * filter)
      filtered_img[y - 1, x - 1] = convolutino_value

  filtered_img = np.clip(filtered_img, 0, 255).astype(np.uint8)

  plt.figure(figsize =(10, 4))
  plt.subplot(1, 2, 1)
  plt.title("Bez obrbki")
  plt.imshow(img, cmap='gray')


  plt.subplot(1, 2, 1)
  plt.title("Naoony filtr")
  plt.imshow(filtered_img,cmap='gray')
  plt.show()
```

```python
apply_filter(img_gray, np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
]))
```

```python
apply_filter(img_gray, np.array([
    [ 2,  1,  0],
    [ 1,  0, -1],
    [ 0, -1, -2]
]))
```

```python
apply_filter(img_gray, np.array([
    [-2, -1, 0],
    [-1,  1, 1],
    [ 0,  1, 2]
]))
```

```python
apply_filter(img_gray, np.array([
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9]
]))
```

```python
apply_filter(img_gray, np.array([
    [1/16, 2/16, 1/16],
    [2/16, 4/16, 2/16],
    [1/16, 2/16, 1/16]
]))
```
