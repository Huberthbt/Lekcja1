```python
import matplotlib.pyplot as plt
import numpy as np

#xpoints = [-3, -2, -1, 0, 1, 2, 3]
xpoints = np.arange(-4,4,0.1)

ypoints = [x ** 2 - 3 for x in xpoints]

plt.plot(xpoints, ypoints,  c = 'purple', marker = 'p', linestyle='--',linewidth=2,label="y = x^2 - 3"     )

plt.title("Wykres funkcji kwadratowej")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.grid()
plt.show()
```
